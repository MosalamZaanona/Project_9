```python
# Importing the libraries we will use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Choosing our topic and finding data:
#### We have picked a kaggle data set that captures Neteflix shows from 2015 till 2021 with several information for each show. The interesting part about this data set is that it has a new attribute called the hidden gem which gives the show a rating out of 10. The lower the review count and the higher the user rating for the show leads to a higher hidden gem score.
Here is the link for the data set on kaggle: https://www.kaggle.com/datasets/ashishgup/netflix-rotten-tomatoes-metacritic-imdb

```python
netflix = pd.read_csv('netflix-rotten-tomatoes-metacritic-imdb.csv')
netflix
```

### Our Analysis:
As we mentioned, this data set has a special attribute called "hidden gem". We want to know what the special thing about hidden gems is. Here are some interesting questions to ask:

- What are the attributes of a show with a high hidden gem score?
- Which languages compose the majority of the shows that have a high hidden gem score?
- Are such shows tend to be not nominated for awards or not getting lots of IMDb reviews?
- Is that related to the Boxoffice income of the shows?




### Analyzing the language data for each show


First, we need to set a rule for the shows that we want to look at. We'll look at the shows with gem score greater than 7. We found that to be a good range for shows that can be considered hidden gems.

```python
hidden_gems = netflix[netflix["Hidden Gem Score"] >= 7]
hidden_gems
```

We want to see the top 10 most common languages for hidden gem shows.

```python
language = hidden_gems['Languages']
language_counts = language.value_counts()
language_counts[0:10]
```

- English is the first on the list which makes sense because the majority of Neflix shows are in English.
- There is a good number of Japanese shows. These could be Anime shows which tend to be higly rated, but not as popular as hollywood or such productions.
- Korean shows are also good examples for hidden gems. They tend to be great, but just a few people know about them.

### Awards Nominations

```python
nominations = hidden_gems['Awards Nominated For']
nominations.value_counts()
```

Based on the results, we can tell that most of hidden gems tend to be nominated for few awards or not nominated at all. However, some of them are nominated for high number of awards, such as 108 and 75 awards. So, number of awards is not a decisive factor for the hidden gems.

```python
#Let's figure out which show had 108 nominations for awards. Did it win an award out of these nominations?
hidden_gems[hidden_gems['Awards Nominated For']==108]
```

```python
hidden_gems[hidden_gems['Title']=='Odd Squad']['Awards Received']
```

This show has won 41 awards out the 108 that was nominated for. Despite winning all of these awards, the show has only 867 IMDb reviews. That's a little number in comparison to the popular shows that get similar number of awards. This is why it has a relatively high hidden gem score of 7.8

Now that we understand the importance of reviews, we want to see the average number of reviews of hidden gems. Then, we will compare it to the avarage number of reviews of shows that have a hidden gem score below 7 (which are relatively non-hidden gems).

```python
reviews = hidden_gems['IMDb Votes']
reviews.mean()
```

```python
#Creating a data set for non-hidden gems
non_hidden_gems = netflix[netflix["Hidden Gem Score"] < 7]
reviews_2 = non_hidden_gems['IMDb Votes']
reviews_2.mean()
```

That is a huge differece that worths investigating.

We want to devide our original data set into 10 groups according to the hidden gem score.

```python
def score(x):
    if x <= 1:
        return '0-1 score'
    elif x <=2:
        return '1-2 score'
    elif x <=3:
        return '2-3 score'
    elif x <=4:
        return '3-4 score'
    elif x <=5:
        return '4-5 score'
    elif x <=6:
        return '5-6 score'
    elif x <=7:
        return '6-7 score'
    elif x <=8:
        return '7-8 score'
    elif x <=9:
        return '8-9 score'
    elif x <=10:
        return '9-10 score'
    else:
        return "Unrated"
    
netflix['gem score range'] = netflix['Hidden Gem Score'].apply(score)
netflix['IMDb Votes'].fillna(value=0, inplace = True)
netflix
```

### Visualization for number of reviews for each range of hidden gem scores

```python
#Let's clean our data before graphing

#First, we need to get rid of data with no hidden gem score
netflix_filtered = netflix[netflix["gem score range"] != "Unrated"]

#Sorting our data according to their gem score
netflix_filtered.sort_values(by = 'Hidden Gem Score', inplace = True)

sns.stripplot(data=netflix_filtered, x="gem score range", y="IMDb Votes", color = "purple")

plt.xticks(rotation= 310)
#We use this scale as it worked the best for our data
plt.yscale("log")


plt.show()
```

From the graph, we can clearly see the difference between number of reviews for high gem scores and low ones. For example, shows with a range of gem score from 3 to 4 have about ten thousands to one million reviews. However, shows falling in the score range from 7 to 8 have about ten to fifty thousand reviews.

This illustrates the relationship between gem score and number of reviews. The higher the hidden gem score is, the lower the number of reviews will be.

### Boxoffice income

Now, we'll check if the box income has something to do with the gem socre.

```python
#add a column for income in a number format
netflix_filtered["Boxoffice_number"] =netflix_filtered["Boxoffice"].replace('[\$,]', '', regex=True).astype(float)

#finding the average boxincome for each range of gem scores
ranges = ['0-1 score', '1-2 score', '2-3 score', '3-4 score', '4-5 score', '5-6 score', '6-7 score', '7-8 score', 
          '8-9 score', '9-10 score']
average_income = []
for i in ranges:
    income = np.mean(netflix_filtered[netflix_filtered['gem score range'] == i]['Boxoffice_number'])
    average_income.append(income)

    
average_income
```

This doesn't show a direct correlation with gem score. But, generally, gem scores below 5 have much more income than those with score above 5.

## New analysis using directors info

We are interested to see whether some directors tend to produce shows with specific gem score range or not.

We can't work on all directors as that will be too much. We will focus on directors with more than 10 shows as this will provide enough amount of data to understand the director's style.

We will also use our filtered data set to benefit from the cleaning that we have done.

```python
Directors = netflix_filtered['Director'].value_counts()
Directors_new = []
for x in Directors.index:
    if Directors[x] >= 10:
        Directors_new.append(x)
Directors_new
```

```python
netflix_dr = netflix_filtered[netflix_filtered['Director'].isin(Directors_new)]
```

```python
netflix_dr
```

```python
#We want to export that data set to use to make our visualization using Flourish.
netflix_dr.to_csv('Netflix Directors.csv')
```

We've used something called radial tree graphing. It is a perfect way to show directors with the distribution of their shows in terms of gem score range.

We can conclude for example that Marcus Raboy mostly produces shows with high gem scores as he has 12 shows with a score in range 7-8. He also has 4 shows in the range of 8-9. These are 16 shows with a high gem score while he has only one show with a low score.

This means Marcus produces good shows, but he is not famous.
